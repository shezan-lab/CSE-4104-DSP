```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate an analog signal (sine wave)
def generate_analog_signal(duration=1, sampling_rate=1000, frequency=5, amplitude=1):
    t = np.linspace(0, duration, int(duration * sampling_rate), endpoint=False)
    analog_signal = amplitude * np.sin(2 * np.pi * frequency * t)
    return t, analog_signal

# Function to sample the signal
def sample_signal(analog_signal, sampling_rate):
    sampling_indices = np.arange(0, len(analog_signal), sampling_rate)
    sampled_signal = analog_signal[sampling_indices]
    return sampled_signal

# Function for quantization
def quantize_signal(signal, num_bits=8):
    quantization_levels = 2 ** num_bits
    quantized_signal = np.round(signal * (quantization_levels - 1)) / (quantization_levels - 1)
    return quantized_signal

# Function for unit impulse
def unit_impulse(n):
    return 1 if n == 0 else 0

# Function for unit step
def unit_step(n):
    return np.heaviside(n, 1)

# Function for unit ramp
def unit_ramp(n):
    return np.maximum(n, 0)

# Function for shifting
def shift_signal(signal, shift):
    return np.roll(signal, shift)

# Function for scaling
def scale_signal(signal, scale):
    return signal * scale

# Function for reversal
def reverse_signal(signal):
    return np.flip(signal)

# Generate analog signal
duration = 1
sampling_rate = 1000
frequency = 5
amplitude = 1
t, analog_signal = generate_analog_signal(duration, sampling_rate, frequency, amplitude)

# Plot analog signal
plt.figure(figsize=(10, 6))
plt.subplot(2, 3, 1)
plt.plot(t, analog_signal)
plt.title('Analog Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Sample and quantize the signal
sampled_signal = sample_signal(analog_signal, sampling_rate)
quantized_signal = quantize_signal(sampled_signal)

# Plot sampled signal
plt.subplot(2, 3, 2)
plt.stem(sampled_signal)
plt.title('Sampled Signal')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')

# Plot quantized signal
plt.subplot(2, 3, 3)
plt.stem(quantized_signal)
plt.title('Quantized Signal')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')

# Plot unit impulse
n = np.arange(-10, 11)
impulse = [unit_impulse(i) for i in n]
plt.subplot(2, 3, 4)
plt.stem(n, impulse)
plt.title('Unit Impulse')
plt.xlabel('n')
plt.ylabel('Amplitude')

# Plot unit step
step = [unit_step(i) for i in n]
plt.subplot(2, 3, 5)
plt.stem(n, step)
plt.title('Unit Step')
plt.xlabel('n')
plt.ylabel('Amplitude')

# Plot unit ramp
ramp = [unit_ramp(i) for i in n]
plt.subplot(2, 3, 6)
plt.stem(n, ramp)
plt.title('Unit Ramp')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

# Shifting, scaling, and reversal examples
shifted_signal = shift_signal(analog_signal, 100)
scaled_signal = scale_signal(analog_signal, 2)
reversed_signal = reverse_signal(analog_signal)

# Plot shifted, scaled, and reversed signals
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, analog_signal)
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(t, shifted_signal)
plt.title('Shifted Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(t, scaled_signal)
plt.title('Scaled Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 3))

plt.subplot(1, 2, 1)
plt.plot(t, analog_signal)
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(1, 2, 2)
plt.plot(t, reversed_signal)
plt.title('Reversed Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
