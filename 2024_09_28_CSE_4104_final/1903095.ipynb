### You are given the following signal:
```
x = {2<-, 4, 6, 8, 10, 12}
```
Find out where 
```
y(n) = 0.2 * x * ( -2 * n - 4 )
```
You need to show the changes graphically after each step

````python
import numpy as np
import matplotlib.pyplot as plt

# 1. Original signal: x(n)
n = np.arange(0, 6) 
x_n = np.array([2, 4, 6, 8, 10, 12])

PLT_GRID = False
````

````python

# Plot original signal
plt.stem(n, x_n)
plt.title("Original Signal: x(n)")
plt.xlabel("n")
plt.ylabel("x(n)")
plt.grid(PLT_GRID)
plt.show()
````

````python

# Time reversal: x(-n)
n_reversed = -n
plt.stem(n_reversed, x_n)
plt.title("Time Reversed Signal: x(-n)")
plt.xlabel("n")
plt.ylabel("x(-n)")
plt.grid(PLT_GRID)
plt.show()
````

````python

# Time scaling: x(-2n)
n_scaled = -2 * n
plt.stem(n_scaled, x_n)
plt.title("Time Scaled Signal: x(-2n)")
plt.xlabel("n")
plt.ylabel("x(-2n)")
plt.grid(PLT_GRID)
plt.show()
````

````python

# Time shift: x(-2n - 4)
n_shifted = n_scaled - 4
plt.stem(n_shifted, x_n)
plt.title("Time Shifted Signal: x(-2n - 4)")
plt.xlabel("n")
plt.ylabel("x(-2n - 4)")
plt.grid(PLT_GRID)
plt.show()
````

````python

# Amplitude scaling: y(n) = 0.5 * x(-2n - 4)
y_n = 0.5 * x_n
plt.stem(n_shifted, y_n)
plt.title("Final Signal: y(n) = 0.5 * x(-2n - 4)")
plt.xlabel("n")
plt.ylabel("y(n)")
plt.grid(PLT_GRID)
plt.show()
````
